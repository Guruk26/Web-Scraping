```python
import requests
import selenium
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
import warnings
warnings.filterwarnings('ignore')

def get_url(position, location, min_exp):
    position = position.replace(' ', '+')
    location = location.replace(' ', '+')
    template = 'https://www.shine.com/job-search/{}-jobs-in-{}?q={}&loc={}&minexp={}'
    url = template.format(position, location, position, location, min_exp)
    return url

def find_job(card):
    # Fetching job_name
    try:
        job_name = card.find('a').text
    except:
        job_name = 'Not Mentioned'
        
    # Fetching company_name
    try:
        company_name = card.find('div','jobCard_jobCard_cName__mYnow').text
    except:
        company_name = 'Not Mentioned'
        
    # Fetching company_location
    try:
        company_location = card.find('div','jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2').text
    except:
        company_location = 'Not Mentioned'
        
    # Fetching experience_required
    try:
        experience_required = card.find('div','jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t').text
    except:
        experience_required = 'Not Mentioned'
        
    # Fetching job_description
    try:
        job_description = card.find('ul','jobCard_jobCard_jobDetail__jD82J').text
    except:
        job_description = 'Not Mentioned'
        
    # Fetching job_apply_url
    try:
        job_apply_url = card.find('a')
        job_apply_url = 'https://www.shine.com' + job_apply_url['href']
    except:
        job_apply_url = 'Not Mentioned'
    
    all_job_detail = (job_name,company_name,company_location,experience_required,job_description,job_apply_url)
    return all_job_detail

def main(position, location, min_exp):
    url = get_url('Java','Pune','2')
    
    driver = webdriver.Firefox()
    driver.get(url)
    driver.maximize_window()
    time.sleep(10)
    
    soup = BeautifulSoup(driver.page_source,'html.parser')
    
    cards = soup.find_all('div','jobCard_jobCard__jjUmu white-box-border jobCard')
    
    all_jobs = []
    for i in cards:
        temp = find_job(i)
        all_jobs.append(temp)
        
    # saving this data to a csv file
    col = ['job_name','company_name','company_location','experience_required','job_description','job_apply_url']
    Shine_jobs_data = pd.DataFrame(all_jobs,columns=col)
    Shine_jobs_data.to_csv('C:\Users\ADMIN\Desktop\Project',encoding='utf-8')

main('Java','Pune','2')
```
